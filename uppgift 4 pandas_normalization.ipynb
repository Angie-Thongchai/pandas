```python
#import methode
import pandas as pd
import numpy as np
import math
```

```python
#import csv file and display
data = pd.read_csv('Automobile.csv', index_col=0)
data.head()
```

```python
#calculate wheel min-max normalization and z-score
wheel = data[['wheel-base']]
wheel['min-man normalization']=(wheel['wheel-base'] -wheel['wheel-base'].mean())/(wheel['wheel-base'].max()-wheel['wheel-base'].min())
wheel['z-score']=(wheel['wheel-base']-wheel['wheel-base'].mean())/(wheel['wheel-base'].std(ddof=0))
wheel
```

```python
#calculate length min-max normalization and z-score
length=data[['length']]
length['min-man normalization']=(length['length'] -length['length'].mean())/(length['length'].max()-length['length'].min())
length['z-score']=(length['length']-length['length'].mean())/(length['length'].std(ddof=0))
length
```

```python
#calculate power min-max normalization and z-score
power=data[['horsepower']]
power['min-man normalization']=(power['horsepower'] -power['horsepower'].mean())/(power['horsepower'].max()-power['horsepower'].min())
power['z-score']=(power['horsepower']-power['horsepower'].mean())/(power['horsepower'].std(ddof=0))
power
```

```python
#calculate average min-max normalization and z-score
average=data[['average-mileage']]
average['min-man normalization']=(average['average-mileage'] -average['average-mileage'].mean())/(average['average-mileage'].max()-average['average-mileage'].min())
average['z-score']=(average['average-mileage']-average['average-mileage'].mean())/(average['average-mileage'].std(ddof=0))
average
```

```python
#calculate price min-max normalization and z-score
price=data[['price']]
price['min-man normalization']=(price['price'] -price['price'].mean())/(price['price'].max()-price['price'].min())
price['z-score']=(price['price']-price['price'].mean())/(price['price'].std(ddof=0))
price
```

```python
#pack every dataframe
df =data[['company']]
result=pd.concat([df,wheel,length,power,average,price],axis=1)
result
```

```python
#export dataframe as csv's file
result.to_csv('newauto.csv')
```
