```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Automobile.csv', index_col=0)
data.head(5)
```

```python
data.tail(5)
```

```python
data_select = data[['company','price']]
```

```python
data_select
```

```python
data_sel = data_select.groupby(['company'], as_index = False)[['price']].max()
data_sel
```

```python
toyota = data_select.loc[data_select['company'].isin(['toyota'])]
toyota
```

```python
count_data = data.groupby(['company'], as_index = False).agg({'price':['max','count']})
count_data
```

```python
rank['price_rank'] = data_sel['price'].rank(ascending = False)
rank = pd.DataFrame(rank['price_rank'])
rank
```

```python
New_col =pd.concat([data_sel,rank], axis = 1)
New_col.drop(columns=['price'])
```

```python
eurocar = New_col[['company']]
carlist = eurocar.loc[eurocar['company'].isin(['audi','bmw','jaguar','mercedes-benz','porsche','volkswagen','volvo'])]
carlist.rename(columns={'company':'europe_car'})
```
